# Error analysis 


It can be helpful to understand why certain transcription fail. 

one way is to compute Word Error Rate and Character Error rate for each example and identify the failed transcription 



```python
import pandas as pd 
from jiwer import wer, cer

df = pd.read_excel('test_transcription_3.xlsx')

df.head()
```

```python
df["wer"] = df.apply(lambda row: wer(row["text_ref"], row["text_tra"]), axis=1)
df["cer"] = df.apply(lambda row: cer(row["text_ref"], row["text_tra"]), axis=1)


df.head()
```

```python
df.to_excel('transcriptions_error_analysis.xlsx', index=False)
```

```python
mean_wer = df["wer"].mean()
mean_cer = df["cer"].mean()

print("Average WER:", mean_wer)
print("Average CER:", mean_cer)
```

```python
q1 = df.iloc[4].to_dict()
```

```python
import regex as re

def normalize(text):
    text = text.lower()
    text = re.sub(r"[^\p{L}'\s]", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text
```

```python
wer(q1.get('text_ref'), q1.get('text_tra'))
```

```python
wer(normalize(q1.get('text_ref')), normalize(q1.get('text_tra')))
```

```python
wer([normalize(x) for x in df['text_ref'] ], [normalize(x) for x in df['text_tra'] ])
```
